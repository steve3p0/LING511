### ✅ Tree #4 <font color='green'>CORRECT</font>

Here is the correct syntax tree with the null NP done manually.

```python
sentence = "The buffalo looked too young to be dangerous"
print(sentence)

correct_t = inspect.cleandoc("""
    (TP(NP (D The) (N buffalo))(VP (V looked) (AdjP  (AdvP (Deg too))  (Adj young)  (TP (NP {tree.EMPTY_SET}) (VP (P to) (V be) (AdjP (Adj dangerous)))))))
    """)
correct_t = inspect.cleandoc("""
    (TP
        (NP 
            (D The)  
            (N buffalo)
        )
        (VP 
            (V looked) 
            (AdjP  
                (AdvP (Deg too))  
                (Adj young)  
                (TP 
                    (NP ∅) 
                    (VP 
                        (P to) 
                        (V be) 
                        (AdjP 
                            (Adj dangerous)
                        )
                    )
                )
            )
        )
    )""")

t = nltk.Tree.fromstring(correct_t)
display(t)
```
