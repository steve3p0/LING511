```python
sentence = "The buffalo waded into the river and swam"
t = tree.parse_sentence(sentence)
display(t)
```

### ✅ Tree #4 <font color='green'>CORRECT</font>
Here is the correct version with the correct POS tag for <i><b>down</b><i> and <i><b>out</b><i> as prepositions.

```python
correct_t = inspect.cleandoc("""
(TP
  (NP (D The) (N buffalo))
  (VP
    (VP (V waded) (PP (P into) (NP (D the) (N river))))
    (VP (Conj and) (N swam))))""")

t = nltk.Tree.fromstring(correct_t)
display(t)
```

### ✅ Tree #6 <font color='green'>CORRECT</font>
Here is the correct version with the correct POS tag for <i><b>down</b><i> and <i><b>out</b><i> as prepositions.

```python
correct_t = inspect.cleandoc("""
    (TP
      (NP (N He))
      (VP
        (V tumbled)
        (PP (P down) (NP (D the) (N river)))
        (PP (P out) (PP (P of) (NP (N control))))))""")

t = nltk.Tree.fromstring(correct_t)
display(t)
```
